```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Dropout
from numpy.random import random,randint
```

```python
#defining the training data
```

```python
sujantrain=random((1000,20))
sujantrain
```

```python
janutrain=randint(20,size=(1000,1))
janutrain
```

```python

#defining the test data
```

```python
sujantest=random((100,20))
```

```python
janutest=randint(20,size=(100,1))
```

```python
import keras
janutrain=keras.utils.to_categorical(janutrain,num_classes=20)
```

```python
janutest=keras.utils.to_categorical(janutest,num_classes=20)
```

#defining the model

```python
model=Sequential()
model.add(Dense(64,input_dim=20,activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(64,activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(20,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(sujantrain,janutrain,epochs=150,batch_size=150)
```

```python
model.evaluate(sujantest,janutest,batch_size=32)
```
