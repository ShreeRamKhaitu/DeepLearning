#importing Libraries

```python
import tensorflow # backend
import keras # frontend
from keras.layers import Dense # Dense layer 
from keras.optimizers import SGD,RMSprop,Adam # helps to define SGD,...
from keras.models import Sequential # Sequential Model
import matplotlib.pyplot as p #helps in plottng the graph
import numpy as np # for array operation
```

```python

#defining Datasets
from numpy.random import random,randint
```

```python
x=random((1000,100)) # array of 1000*100
```

```python
x # dispaly x
```

```python
#defining the label
```

```python
l=np.random.randint(10,size=(1000,1)) # array of (1000*1) with 2 class
```

```python
l
```

```python
#COnvert the labels into Categorical (one hot Encoddong)
```

```python
l=keras.utils.to_categorical(l,num_classes=10)
```

```python
# Defining single input model for Categorical classification
```

```python
model=Sequential()
```

```python
model.add(Dense(32,activation='relu',input_dim=100)) 
#Dense layers with 32 hidden units with activation as relu
# input dimension is 100 i.e input with 100 coiumn
```

```python
model.add(Dense(10,activation='softmax'))
# 10 refersr the ccategorical classification using softmax activation function
```

```python
#compile the model
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# training
model.fit(x,l,epochs=150,batch_size=32)
```

```python
len(model.layers)
```

```python
model.summary()
```

```python
s=model.evaluate(x,l,batch_size=32,verbose=0)
print(s[0],'\n')
print(s[1])
```

```python
model.predict(x,batch_size=32,verbose=0)
```
