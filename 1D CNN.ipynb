```python
#importing Libraries
```

```python
import keras
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,Embedding
from keras.layers import Convolution1D as conv1d
from keras.layers import GlobalMaxPooling1D
```

```python
from keras.datasets import imdb
```

```python

#loading the imdb datasets
```

```python
(xtrain,ytrain),(xtest,ytest)=imdb.load_data(num_words=5000)
```

```python
len(xtrain)
```

```python
len(xtest)
```

```python
#padding the sequence 400  times
```

```python
xtrain=sequence.pad_sequences(xtrain,maxlen=400)
```

```python
xtest=sequence.pad_sequences(xtest,maxlen=400)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
#defining the model
```

```python
model=Sequential()
```

```python
model.add(Embedding(5000,50,input_length=400))
```

```python
model.add(Dropout(0.4))
```

```python
#Adding 1D convolution layer
```

```python
model.add(conv1d(250,3,padding='valid',activation='relu',strides=1))
```

```python
model.add(GlobalMaxPooling1D())
```

```python
model.add(Dense(250))
```

```python
model.add(Activation('relu'))
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,batch_size=32,epochs=3,
          validation_data=(xtest,ytest))
```

```python
keras.utils.plot_model(model,'cnn1.png',show_shapes=True)
```
