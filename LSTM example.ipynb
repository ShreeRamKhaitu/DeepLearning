```python
maxlen=30 # maximum length for dividing the sentence
step=3
sentences=[]
next_word=[]
sentence1=[]
list_word=[]
sentence2=[]
```

```python
list_word=text.split() # defining the all word array
```

```python
# chunking
```

```python
for i in range(0,len(list_word)-30,step):
    sentence2=' '.join(list_word[i:i+maxlen])
    sentences.append(sentence2)
    next_word.append(list_word[i+maxlen])
```

```python
#vectorization
```

```python
x=np.zeros((len(sentences),maxlen,len(words)),dtype=np.bool)
```

```python
y=np.zeros((len(sentences),len(words)),dtype=np.bool)
```

```python
# making the dictionary of the sentences
```

```python
for i,sentence in enumerate(sentences):
    for t,word in enumerate(sentence.split()):
        x[i,t,word_indices[word]]=1
    y[i,word_indices[next_word[i]]]=1
```

```python
#defining Model
```

```python
model=Sequential()
```

```python
model.add(LSTM(512,return_sequences=True,
               input_shape=(maxlen,len(words))))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(LSTM(512,return_sequences=False))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Dense(len(words)))
```

```python
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam')
```

```python
#helper function
```

```python
def sample(a,temp=1.0):
    a=np.log(a)/temp
    a=np.exp(a)/np.sum(np.exp(a))
    return np.argmax(np.random.multinomial(1,a,1))
```

```python
# train the model, output generated text after each iteration
```

```python
for iteration in range(1,2):
    print()
    print('_'*50)
    print('Iteration',iteration)
    model.fit(x,y,batch_size=128,epochs=2)
    model.save_weights('srk_weights',overwrite=True)
    
    start_index=random.randint(0,len(list_word)-maxlen-1)
    
    for diversity in [1.0]:
        print()
        print('-----diversity',diversity)
        generated=' '
        sentence=list_word[start_index:start_index+maxlen]
        generated+=' '.join(sentence)
        print('----Generating with seed:"',sentence,'"')
        print()
        for i in range(128):
            x=np.zeros((1,maxlen,len(words)))
            for t,word in enumerate(sentence):
                x[0,t,word_indices[word]]=1
            preds=model.predict(x,verbose=0)[0]
            next_index=sample(preds,diversity)
            next_word=indices_word[next_index]
            generated+=next_word
            del sentence[0]
            sentence.append(next_word)
            sys.stdout.write(' ')
            sys.stdout.write(next_word)
            sys.stdout.flush()
        print()
    
    
```
