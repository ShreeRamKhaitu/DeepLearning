```python
#importing libraries
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import numpy as np 
```

```python
#defining x
```

```python
x=np.random.random((500,1))
```

```python
#defining y
```

```python
y=np.dot(x,2.)+4 # y=2x + 4
```

```python
x
```

```python
y
```

```python
# defining  model
```

```python
nabin=Sequential()
```

```python
nabin.add(Dense(1,input_shape=(1,)))
# 1 for only one output and input shape= 1 becose input has 1 column
```

```python
#Compilling
```

```python
nabin.compile(optimizer=SGD(lr=0.001),loss='mse')
```

```python
nabin.get_weights() # weight befor training
```

```python
nabin.evaluate(x,y) # evaluates before training
```

```python
# Training
```

```python
nabin.fit(x,y,epochs=1000,batch_size=100)
```

```python
nabin.evaluate(x,y,verbose=0)
```
