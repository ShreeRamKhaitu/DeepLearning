#importing Libraries

```python
import tensorflow # backend
import keras # frontend
from keras.layers import Dense # Dense layer 
from keras.optimizers import SGD,RMSprop,Adam # helps to define SGD,...
from keras.models import Sequential # Sequential Model
import matplotlib.pyplot as p #helps in plottng the graph
import numpy as np # for array operation
```

```python

#defining Datasets
from numpy.random import random,randint
```

```python
x=random((1000,100)) # array of 1000*100
```

```python
x # dispaly x
```

```python
#defining the label
```

```python
l=np.random.randint(2,size=(1000,1)) # array of (1000*1) with 2 class
```

```python
l
```

```python
# Defining single input model for binary classification
```

```python
model=Sequential()
```

```python
model.add(Dense(32,activation='relu',input_dim=100)) 
#Dense layers with 32 hidden units with activation as relu
# input dimension is 100 i.e input with 100 coiumn
```

```python
model.add(Dense(1,activation='sigmoid'))
# 1 refersr the binary classification using sigmoid activation function
```

```python
#compile the model
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# training
model.fit(x,l,epochs=150,batch_size=32)
```

```python
len(model.layers)
```

```python
model.summary()
```

```python
model.evaluate(x,l,batch_size=32,verbose=0)
```

```python
model.predict(x,batch_size=32,verbose=0)
```
